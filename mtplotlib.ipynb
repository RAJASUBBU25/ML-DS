```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3],[2,4,6])

plt.title('Our first graph', fontdict={'fontname':'Comic Sans MS'})
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
plt.plot([1,2,3],[2,4,6],"X")
```

```python
plt.plot([1,2,3],[2,4,6],"X"'r') #Colour
```

```python
ypts = np.array([3,8,1,12])
xpts = np.array([5,6,7,12])
plt.plot(xpts,ypts, marker='*', ms=20, color='black') #ms(marker size)
plt.show()
```

```python
ypts = np.array([3,8,1,12])
xpts = np.array([5,6,7,12])
plt.plot(xpts,ypts, marker='o', ms=20,mec='green', mfc='pink',color='k') #ms(marker size)
plt.show()
```

```python
plt.plot(xpts,ypts,linestyle='dashed', marker='o', ms=20,mec='green', mfc='pink',color='k')
```

```python
plt.plot(xpts,ypts,linestyle= 'dashed',linewidth=5, marker='o', ms=20,mec='green', mfc='pink',color='k')
```

```python
plt.plot(xpts,ypts,linestyle= 'dotted',linewidth=5, marker='o', ms=20,mec='green', mfc='pink',color='k')
```

```python
#If we dont specify x values it takes 0,1,2,3,4........
plt.plot([3,8,1,10])
plt.plot([6,2,5,11,10,12])
plt.show()
```

```python
plt.plot([1,2,3],[2,4,6])

plt.title('Our first graph',loc="left")
plt.xlabel('X axis',loc="right")
plt.ylabel('Y axis',loc="top")

plt.grid(color='green')

plt.show()
```

```python
plt.plot([1,2,3],[2,4,6])

plt.title('Our first graph',loc="left")
plt.xlabel('X axis',loc="right")
plt.ylabel('Y axis',loc="top")
plt.grid(color='green', linestyle ='dotted')

plt.show()
```
